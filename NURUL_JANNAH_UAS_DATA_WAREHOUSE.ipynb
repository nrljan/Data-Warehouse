### 14. Menampilkan nama student yang paling rajin masuk

```python
import pandas as pd
import pandas.io.sql as sqlio
import psycopg2
from psycopg2.extensions import ISOLATION_LEVEL_AUTOCOMMIT

conn = psycopg2.connect(database = "db_uas", user = "postgres", password = "NurulNurul", host = "localhost", port = "5432")
cur = conn.cursor()

sql = "SELECT * FROM TrD_Absensi;"
dat = sqlio.read_sql_query(sql,conn)
trans = dat.transpose()
id_murid_rajin = []
besar = 0
for i in range(10):
    a = trans[i].str.count('H')
    if sum (a[2:len(a)])>besar :
        besar = sum(a[2:len(a)])
        id_murid_rajin.clear()
        id_murid_rajin.append(dat['student_id'][i])
    elif sum(a[2:14]) == besar :
        id_murid_rajin.append(dat['student_id'][i])
id_murid_rajin = tuple(id_murid_rajin)
cur.execute(f"SELECT Student_name FROM Ms_Student WHERE student_id IN {id_murid_rajin}")
print("Murid yang paling rajin masuk adalah", ','.join([i[0] for i in cur.fetchall()]))
```

### 15. Menampilkan nama student yang paling malas masuk

### 18. Menampilkan nama program dan nama student yang mendapatkan hasil Note Excellent

```python
import psycopg2

conn = psycopg2.connect(database = "db_uas", user = "postgres", password = "NurulNurul", host = "localhost", port = "5432")
cur = conn.cursor()

cur.execute('''
    SELECT Student_name, Program_name FROM Ms_Program, Ms_Student, Tr_Result
    WHERE Note = 'EXCELLENT' AND Ms_Program.Program_id = Tr_Result.Program_id
    AND Ms_Student.Student_id = Tr_Result.Student_id
''')

data = cur.fetchall()
nama = [x[0] for x in data]
program = [x[1] for x in data]
for x in range(len(data)) :
    print(f"Murid yang mendapatkan note Excellent adalah {nama[x]} pada program {program[x]}")
```
